```
def square(x):
  return x ** 2

square(10)
```

```
def sum_to_n(x=int(input())):
  count = 0
  for i in range(x):
    count += i
  return count

sum_to_n()
```

```
def sum_of_squares(x=10):
  count = 0
  for i in range(x):
    count += i ** 2
  return count

sum_of_squares(4)
```

```
def wordMatch(string1, string2):
  words1 = string1.split()
  words2 = string2.split()

  count = 0
  for i in words1:
    if words2.count(i) > 0:
      count += 1

  return count

wordMatch('hello world tree', 'hello uncle tree')
```

```
def AR(p, init_val, list_of_shocks, n):
  y_path = [init_val]
  y = init_val
  for i in range(n):
    y = p * y + list_of_shocks[i]
    y_path.append(round(y, 2))
  return y_path

y = AR(0.9, 0, [0.1, -0.2, 0.4, -0.5, 0.3, 0.2], 6)
y
```

```
def linearModelScore(x, alpha, beta):
  no_of_examples = len(x)
  yhats = [alpha + beta * x[i] for i in range(no_of_examples)]
  return yhats

yhats = linearModelScore([0.4, 0.1, 0.9, 2, 0.5], 0.5, -0.5)
yhats
```
